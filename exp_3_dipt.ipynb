## Grayscale Histogram Equalization

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r"C:\Users\admin\Downloads\parrot.jpg", 0)
```

```python
if img is None:
    print("Error: Image not found!")
    exit()
```

```python
plt.figure(figsize=(8,6))
plt.hist(img.ravel(), 256, [0,256])
plt.title("Histogram of Original Image")
plt.show()
```

```python
equalized_img = cv2.equalizeHist(img)
```

```python
plt.figure(figsize=(10,8))
```

```python
plt.subplot(2,2,1)
plt.imshow(img, cmap='gray')
plt.title("Original Grayscale Image")
plt.axis('off')
```

```python
plt.subplot(2,2,2)
plt.hist(img.ravel(), 256, [0,256])
plt.title("Original Histogram")
```

```python
plt.subplot(2,2,3)
plt.imshow(equalized_img, cmap='gray')
plt.title("Enhanced Image")
plt.axis('off')
```

```python
plt.subplot(2,2,4)
plt.hist(equalized_img.ravel(), 256, [0,256])
plt.title("Enhanced Histogram")
```

```python
plt.tight_layout()
plt.show()
```

## Color Histogram Equalization (HSV Method)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r"C:\Users\admin\Downloads\parrot.jpg")
```

```python
if img is None:
    print("Error: Image not found!")
    exit()
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
colors = ('b','g','r')
plt.figure(figsize=(8,6))
for i, col in enumerate(colors):
    hist = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(hist, color=col)
plt.title("Histogram of Original Color Image")
plt.show()
```

```python
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
h, s, v = cv2.split(hsv)
```

```python
v_eq = cv2.equalizeHist(v)
```

```python
hsv_eq = cv2.merge([h, s, v_eq])
```

```python
img_eq = cv2.cvtColor(hsv_eq, cv2.COLOR_HSV2BGR)
```

```python
img_eq_rgb = cv2.cvtColor(img_eq, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,8))
```

```python
plt.subplot(2,2,1)
plt.imshow(img_rgb)
plt.title("Original Color Image")
plt.axis('off')
```

```python
plt.subplot(2,2,2)
for i, col in enumerate(colors):
    hist = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(hist, color=col)
plt.title("Original Histogram")
```

```python
plt.subplot(2,2,3)
plt.imshow(img_eq_rgb)
plt.title("Enhanced Color Image")
plt.axis('off')
```

```python
plt.subplot(2,2,4)
for i, col in enumerate(colors):
    hist = cv2.calcHist([img_eq],[i],None,[256],[0,256])
    plt.plot(hist, color=col)
plt.title("Enhanced Histogram")
```

```python
plt.tight_layout()
plt.show()
```
